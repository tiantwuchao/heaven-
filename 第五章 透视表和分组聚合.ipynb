```python
# 想看看哈登对阵同一对手在不同主客场下的数据，分类条件为对手和主客场
df.pivot_table(index=['主客场','对手'])
```

```python
df.pivot_table(index=['主客场'],values=['得分','篮板','助攻'])
```

```python
#获取每个队主客场的总得分（在总得分的基础上又进行了对手的分类）
df.pivot_table(index='主客场',columns="胜负")
```

```python
df.pivot_table(index=['主客场','胜负'],values=['得分','篮板','助攻'],aggfunc='sum')
```

```python
df.pivot_table(index='主客场',columns="胜负")
```

```python
df.pivot_table(index='主客场',values='得分',columns='对手',aggfunc='sum')
```

```python
t = df.pivot_table(index='主客场',
               values='得分',
               columns='对手',
               aggfunc='sum',
               fill_value=0,
              margins= True,
             )
```

```python
t.query('对手 == ["灰熊"]')
```

```python
table=pd.pivot_table(df,index=[u'对手',u'胜负'],columns=[u'主客场'],values=[u'得分',u'助攻',u'篮板'],aggfunc=[np.mean],fill_value=0)
```
